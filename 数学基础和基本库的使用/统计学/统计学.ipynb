# 偏度
正态分布的偏度为0，
两侧尾部长度对称。若以bs表示偏度。bs<0称分布具有负偏离，也称左偏态，此时数据位于均值左边的比位于右边的少，直观表现为左边的尾部相对于与右边的尾部要长，因为有少数变量值很小，使曲线左侧尾部拖得很长；bs>0称分布具有正偏离，也称右偏态，此时数据位于均值右边的比位于左边的少，直观表现为右边的尾部相对于与左边的尾部要长，因为有少数变量值很大，使曲线右侧尾部拖得很长；而bs接近0则可认为分布是对称的。

偏度是样本的三阶标准化矩

$$
Skew(X) = E[(\frac {X-\mu}{\sigma})^3] = \frac {k_3}{\sigma^3} = \frac {EX^3 - 3\mu\sigma^2 -\mu^3}{\sigma^3}

$$
三阶中心距

$$
k_3 = E[(X-\mu)^3]
$$

# 峰度
峰度（Kurtosis）与偏度类似，是描述总体中所有取值分布形态陡缓程度的统计量。这个统计量需要与正态分布相比较，峰度为0表示该总体数据分布与正态分布的陡缓程度相同；峰度大于0表示该总体数据分布与正态分布相比较为陡峭，为尖顶峰；峰度小于0表示该总体数据分布与正态分布相比较为平坦，为平顶峰。峰度的绝对值数值越大表示其分布形态的陡缓程度与正态分布的差异程度越大

$$
K = \frac {\frac 1 {n-1} \Sigma_{i=1}^n (x_i - \bar x)^4}{(\frac 1 {n-1} \Sigma_{i=1}^n (x_i - \bar x)^2)^2} -3
$$
