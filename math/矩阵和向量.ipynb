# çŸ©é˜µçš„è¡Œåˆ—å¼ determinant
né˜¶æ–¹é˜µAçš„è¡Œåˆ—å¼å®šä¹‰ä¸º:
$$det(A) = \sum_{\sigma \in S_n}par(\sigma)a_{1\sigma_1}a_{2\sigma_2}\dots a_{n\sigma_n}$$
$S_n$ä¸ºæ‰€æœ‰né˜¶æ’åˆ—çš„é›†åˆ(å…±$\frac {n(n-1)} {2}$),$par(\sigma)$ä¸º-1æˆ–+1,å–å†³äº$\sigma=(\sigma_1,\sigma_1,\dots, \sigma_n)$
ä¸ºå¥‡æ’åˆ—æˆ–å¶æ’åˆ—, å³å…¶ä¸­å‡ºç°**é™åºçš„æ¬¡æ•°**ä¸ºå¥‡æ•°æˆ–å¶æ•°, å¦‚(1, 3, 2)é™åºæ¬¡æ•°ä¸º1, (1, 4, 3, 2)ä¸­é™åºæ¬¡æ•°ä¸º3

# æ ¼æ‹‰å§†çŸ©é˜µ Gram Matrix
